## Obtendo valores em nível de um `MultiIndex`  com `.get_level_values`

```python
bigmac = pd.read_csv("../Dados/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(5)
```

```python
bigmac.index.get_level_values("Date")
bigmac.index.get_level_values(0)
```

```python
bigmac.index.get_level_values("Country")
bigmac.index.get_level_values(1)
```

## Renomeando níveis do índoce com `.set_names`

```python
bigmac = pd.read_csv("../Dados/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(5)
```

```python
bigmac.index.set_names(names= ["Day", "Location"], inplace=True)
bigmac.head(3)
```

```python
bigmac.index.set_names(names="Date", level="Day", inplace=True)
bigmac.head(3)
```

## O método Transpose em Dataframes com `MultiIndex`

```python
bigmac = pd.read_csv("../Dados/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(2)
```

```python
bigmac = bigmac.transpose()
bigmac.head()
```

```python
bigmac.loc[("Price in US Dollars",),("2010-01-01","Brazil"):("2010-01-01","Egypt")]
```

## O método `.swaplevel()`

```python
bigmac = pd.read_csv("../Dados/bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(2)
```

```python
bigmac = bigmac.swaplevel()
bigmac = bigmac.swaplevel(i="Date", j="Country")
```

## O método `.unstack()`

```python
world = pd.read_csv("../Dados/worldstats.csv", index_col = ["country", "year"])
world.head(3)

s = world.stack()
s
```

```python
s.unstack().unstack().unstack()
```

```python
s.unstack(level=1)
s.unstack(level=-2)

s.unstack(level=2)
s.unstack(level=-1)

s.unstack(level=0)
s.unstack(level=-3)
```

```python
s.unstack("year")
s.unstack("country")
```

```python
s.unstack(level = [0,1])
s.unstack(level = ["country", "year"])
s.unstack(level = [1,0])
s.unstack(level = ["year", "country"])
```

```python
s.unstack(level="year", fill_value=0)
```

```python
s = s.unstack(level="year", fill_value=0.0)
s
```

## O métodos `.pivot()` e `.pivot_table()`

```python
sales = pd.read_csv("../Dados/salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales.pivot(index="Date", columns='Salesman', values="Revenue")
```

```python
sales.pivot(index="Date", columns='Salesman', values="Revenue").describe()
```

```python
foods = pd.read_csv("../Dados/foods.csv")
foods["Gender"] = foods["Gender"].astype("category")
foods["City"] = foods["City"].astype("category")
foods["Frequency"] = foods["Frequency"].astype("category")
foods["Item"] = foods["Item"].astype("category")
foods.head(4)
```

```python
foods.pivot_table(values="Spend", index="Gender", aggfunc="max")
foods.pivot_table(values="Spend", index="Gender", aggfunc="sum")
foods.pivot_table(values="Spend", index="Gender", aggfunc="mean")
foods.pivot_table(values="Spend", index="Gender", aggfunc="min")
```

```python
foods.pivot_table(values="Spend", index="Item", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], aggfunc="sum", columns="City")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], aggfunc="count", columns="City")
```
