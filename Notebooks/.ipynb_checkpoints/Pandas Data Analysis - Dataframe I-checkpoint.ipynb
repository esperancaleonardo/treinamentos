### Operações `Broadcast` em `pd.Dataframes`

```python
nba = pd.read_csv('../Dados/nba.csv')
nba.head(2)
```

```python
nba["Age"].add(5)
nba["Age"] + 6

nba["Salary"].sub(200000)
nba["Salary"] - 200000

nba["Weight"].mul(0.453592)
nba["Weight"] * 0.453592
```

```python
nba["Weight (Kg)"] = round(nba["Weight"] * 0.453592, 3)
nba["Salary (Mi)"] = round(nba["Salary"] / 1000000, 2)
nba["Birth Year"] = 2021 - nba["Age"]
nba.head(6)
```

### O método `.astype()` para conversão de tipo de dados

```python
nba = pd.read_csv('../Dados/nba.csv').dropna(how="all")
nba["College"].fillna(value="No College", inplace=True)
nba["Salary"].fillna(value=0, inplace=True)
nba.head(6)
```

```python
nba.dtypes
nba.info()
```

```python
nba["Number"] = nba["Number"].astype(int)
nba["Salary"] = nba["Salary"].astype(int)
nba["Age"] = nba["Age"].astype(int)
nba["Position"] = nba["Position"].astype("category")
nba["Team"] = nba["Team"].astype("category")
nba.head(6)
```

### Ordenando um  `pd.Dataframe` com o método `.sort_values()`

```python
nba = pd.read_csv('../Dados/nba.csv').dropna(how="all")
nba.head(3)
```

```python
nba.sort_values(by="Name")
nba.sort_values(by="Age", ascending=False)
nba.sort_values(by="Salary", ascending=False, na_position="first")

nba.sort_values(by=["Team", "Age", "Name"])
nba.sort_values(by=["Team", "Age", "Name"], ascending=[True, False, False]).head(15)
nba.sort_values(by=["Team", "Age", "Name"], ascending=[True, False, False])
```

### Ordenando o index de um `pd.Dataframe` com o método `.sort_index()` e resetando com `.reset_index()`

```python
nba = pd.read_csv('../Dados/nba.csv').dropna(how="all")
nba.sort_values(by=["Team", "Age", "Name"], ascending=[True, False, False], inplace=True)
nba.head().sort_index().reset_index()
```

### O método `.rank()` para elencar registros

```python
nba = pd.read_csv('../Dados/nba.csv').dropna(how="all")
nba["College"].fillna(value="No College", inplace=True)
nba["Salary"] = nba["Salary"].fillna(value=0).astype(int)
nba.head(4)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(int)
nba.sort_values(by="Salary", ascending=False, inplace=True)
nba.head(15)
```
