### O método `.apply()` em Dataframes

```python
james_bond = pd.read_csv('../Dados/jamesbond.csv', index_col="Film")
james_bond.sort_index(inplace=True)
james_bond.head(1)
```

```python
def str_and_text(number):
    return str(number) + " Millions!!!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]

for col in columns:
    james_bond[col] = james_bond[col].apply(str_and_text)
    
james_bond.head(3)
```

### O método `.apply()` com linhas inteiras do dataframe

```python
james_bond = pd.read_csv('../Dados/jamesbond.csv', index_col="Film")
james_bond.sort_index(inplace=True)
james_bond.head(1)
```

```python
def ranking(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Daniel Craig":
        return "Best Movie Ever!!!"
    elif budget > 40.0:
        return "Good Movie!!"
    else:
        return "No Clue!"
```

```python
james_bond["Ranking"] = james_bond.apply(ranking, axis='columns')
james_bond.head(10)
```
