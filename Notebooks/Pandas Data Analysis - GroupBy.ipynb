## O método `.agg()`

```python
fortune = pd.read_csv("../Dados/fortune1000.csv", index_col="Rank")
sectors = fortune.groupby(by=["Sector", "Industry"])
fortune.head(2)
```

```python
sectors.agg({"Revenue":"max","Profits":"sum","Employees":"mean"})
```

```python
sectors.agg(["size", "sum", "min", "mean", "max", "median"])
```

## Iterando sobre grupos de um `GroupBy`

```python
fortune = pd.read_csv("../Dados/fortune1000.csv", index_col="Rank")
sectors = fortune.groupby(by=["Sector"])
fortune.head(2)
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_rev_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_rev_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
for city, data in cities:
    highest_rev_company_by_city = data.nlargest(1, "Revenue")
    df = df.append(highest_rev_company_by_city)
    
df
```
