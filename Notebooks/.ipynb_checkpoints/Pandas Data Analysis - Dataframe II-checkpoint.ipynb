### Filtrando registros com múltiplas condições

```python
condicao1 = df["Gender"] == "Male"
condicao2 = df["Team"] == "Sales"
condicao3 = df["Salary"] <= 60000


df[condicao1 & condicao2 & condicao3].head(10)
```

```python
condicao1 = df["Gender"] == "Female"
condicao2 = df["Salary"] >= 140000
condicao3 = df["Team"] == "Engineering"

df[condicao1 | condicao3 | condicao2].head(10)
```

```python
condicao1 = df["First Name"] == "Robert"
condicao2 = df["Team"] == "Client Services"
condicao3 = df["Start Date"] >= "2016-06-01"

df[(condicao1 & condicao2) | condicao3].head(2)
```

### O método `.between()`

```python
df = pd.read_csv('../Dados/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head(3)
```

```python
df[df["Salary"].between(60000, 70000)]
df[df["Bonus %"].between(9.5, 11.5)]
```

```python
df[df["Start Date"].between("1995-01-01", "1995-12-31")]
df[df["Last Login Time"].between("8:00AM", "8:15AM")]
```

### O método `.duplicated()` e o método `.drop_duplicates()`

```python
df = pd.read_csv('../Dados/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.sort_values("First Name", inplace=True)
df.head(5)
```

```python
df[df["First Name"].duplicated(keep='first')]
df[df["First Name"].duplicated(keep='last')]
df[~df["First Name"].duplicated(keep=False)]
```

```python
len(df)
```

```python
df.head(5)
```

```python
df_ = df.drop_duplicates(subset=['First Name'], keep='last')
df_.head(3)
```

```python
df_.head(10)
```
